##### Portfolio 1 - Step 1: Using the 60-day rolling window, the daily return data, and the S&P 500 returns, calculate the covariance. Review the last five rows of the covariance of the portfolio.

```python
# Calculate the covariance using a 60-day rolling window 
# Review the last five rows of the covariance data
# YOUR CODE HERE
whales_navs_NO_SPX=whales_navs_df.iloc[:,0:4] 
whales_navs_NO_SPX_cov=whales_navs_NO_SPX.cov()

#print("#########################",whales_navs_NO_SPX_cov)
BRK_df=whales_navs_df.iloc[:,3]
BRK_60_day_rolling=BRK_df.rolling(window=60)
BRK_60_day_rolling_cov=BRK_60_day_rolling.cov()
BRK_60_day_rolling_cov=BRK_60_day_rolling_cov.dropna()


Tiger_df=whales_navs_df.iloc[:,2]
Tiger_60_day_rolling=Tiger_df.rolling(window=60)
Tiger_60_day_rolling_cov=Tiger_60_day_rolling.cov()
Tiger_60_day_rolling_cov=Tiger_60_day_rolling_cov.dropna()

print(Tiger_60_day_rolling_cov.tail())
print(BRK_60_day_rolling_cov.tail())
```

##### Portfolio 1 - Step 2: Calculate the beta of the portfolio. To do that, divide the covariance of the portfolio by the variance of the S&P 500.
